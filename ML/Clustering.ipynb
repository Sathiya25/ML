```r
data <- read.csv('dataset/Mall_Customers.csv')
data
```

```r
data <- data[4:5]
data
```

```r
#Using the elbow method to find the optimal no. of clusters
set.seed(5)
wcss = vector()
for(i in 1:10)
    wcss[i] = sum(kmeans(data,centers = i)$withinss)
wcss
```

```r
plot(wcss, type = 'b')
```

```r
#Fitting K-means to the dataset (centers  = 5 from graph)
kmean  <- kmeans(x = data, centers = 5)
kmean
```

```r
y_kmeans <- kmean$cluster
y_kmeans
```

```r
#visualize
library(cluster)
clusplot(data,y_kmeans,
        #lines = 0,
        #shade = T,
        #color = T,
        #labels = 2,
        #plotchar = F, #makes all cluster symbols to circle
        main = "Cluster of Customers",
        xlab = 'Annual Income',
        ylab = 'Spending Score'
        )
```
