```r
data <- read.csv("dataset/Social_Network_Ads.csv")
```

```r
data
```

```r
data <- data[,c(3:5)]
```

```r
data
```

```r
str(data)
```

```r
#Encoding target feature as factor
data$Purchased <- as.factor(data$Purchased)
str(data)
```

```r
library(caTools)
set.seed(1234)
```

```r
split <- sample.split(Y = data$Purchased, SplitRatio = 3/4)
summary(split)
```

```r
train  <- subset(data, split == T)
train
```

```r
test  <- subset(data, split == F)
test
```

```r
#Future Scaling
train[-3] = scale(train[-3])
train
```

```r
test[-3] = scale(test[-3])
test
```

```r
#fitting logistic regression to the training set
?glm
```

```r
classifier <- glm(formula = Purchased ~ .,
   family = binomial,
   data = train)

summary(classifier)
```

```r
# prediction of test

predict(object = classifier,
       type = "response",
       newdata = test[-3])
```

```r
prob_pred <- predict(object = classifier,
       type = "response",
       newdata = test[-3])
```

```r
y_pred <- ifelse(prob_pred > 0.5, 1, 0)
y_pred
```

```r
#making confusion matrix
table(test[,3],y_pred)
#(TP+TN)/all4
```
